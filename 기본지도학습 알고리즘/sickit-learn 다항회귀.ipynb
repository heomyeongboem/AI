```python
from sklearn import datasets
from sklearn.preprocessing import PolynomialFeatures # PolynomialFeatures 가지도록 전처리 해주는 라이브러리 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error #MSE 라이브러리

import pandas as pd 

#데이타셋 가져오기 -> load_
diabets_dataset = datasets.load_diabetes()

#다항회귀 변환기 정의 ->  polynomialFeatures(숫자)
polynomial_transformer = PolynomialFeatures(2)

#데이터를 다항회귀로 변환 ->  변환기이름.fit_transform(데이타셋.data)
polynomial_data = polynomial_transformer.fit_transform(diabets_dataset.data)

#feature을 다항회귀로 변환 -> 변환기이름.get_feature_names_out(데이타셋.feature_names) 
polynomial_features_name = polynomial_transformer.get_feature_names_out(diabets_dataset.feature_names)

#데이타셋 dataFrame으로 변환 -> X = pd.DataFrame(데이타, columns = 데이타특징 ) 
X = pd.DataFrame(polynomial_data, columns = polynomial_features_name)
y = pd.DataFrame(diabets_dataset.target, columns = ['diabetes'])

#데이터셋 train test로 분리 -> train_test_split(X,y,test_size = 숫자, random_state = 숫자)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 5)

#모델 정의 -> ~()
model = LinearRegression()

#모델 훈련 ->모델.fit(X_train,y_train)
model.fit(X_train, y_train)

#예측값 구하기 -> 모델.predict(X_test)
y_test_prediction = model.predict(X_test)

#오차 구하기(MSE) -> mean_squared_error(y_test,y_prediction)
mse = mean_squared_error(y_test, y_test_prediction)

#RMSE -> mse ** 0.5 
mse ** 0.5 
```
