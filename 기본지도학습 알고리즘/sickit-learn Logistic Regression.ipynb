```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import pandas as pd
```

```python
iris_data = load_iris()
```

```python
print(iris_data.DESCR)
```

```python
X = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
```

```python
X
```

```python
y = pd.DataFrame(iris_data.target, columns= ['class'])
```

```python
y
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size= 0.2, random_state = 5)
```

```python
y_train = y_train.values.ravel()
```

```python
model = LogisticRegression(solver = 'saga', max_iter = 2000) # solver = 'saga' 는 최적화 함수 , max_iter = 2000는 최대반복 2000번 
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test) #분류한것중에 맞은것의 갯수를 확률로 나타냄 
```

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import pandas as pd 
wine_data = load_wine()

X = pd.DataFrame(wine_data.data, columns = wine_data.feature_names)
y =pd.DataFrame(wine_data.target,columns = ['Y/N'])

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 5)
y_train = y_train.values.ravel() #error 방지

logistic_model = LogisticRegression(solver = 'saga', max_iter = 7500)

logistic_model.fit(X_train,y_train) # 훈련

# 예측값 (기호에 맞게) -> model.predict(X_test)
y_test_predict = logistic_model.predict(X_test) 

# 로지스틱 회귀 모델의 성능 확인 (정확성 %를 리턴함) -> model.score(X_test,y_test)
score = logistic_model.score(X_test,y_test)
y_test_predict, score
```
