```python
import pandas as pd 

titanic_df = pd.read_csv('titanic.csv')
```

```python
titanic_df.head()
```

```python
titanic_sex_embarked = titanic_df[['Sex','Embarked']]
```

```python
titanic_sex_embarked.head()
```

##전체 원 핫 인코딩
pd.get_dummies(데이터)

```python
#one-hot encoding 하기  ->pd.get_dummies(데이터프레임)
one_hot_encoded_df = pd.get_dummies(titanic_sex_embarked)
```

```python
one_hot_encoded_df.head() # Sex , Embarked가 각 카테고리별로 이루어진 열로 만들어진 것을 볼 수 있다. 
```

##원래 데이터셋에서 원하는 열만 one-hot-encoding 
pd.get_dummies(data = 데이터, columns= ['원핫인코딩할열'])

```python
one_hot_encoded_df = pd.get_dummies(data = titanic_df, columns= ['Sex','Embarked'])
one_hot_encoded_df.head() # 새로운 열 Sex_female	Sex_male	Embarked_C	Embarked_Q	Embarked_S 이 생긴 것을 볼 수 있다.
```

```python
import pandas as pd 
gender_df = pd.read_csv('gender.csv')

#gender_df 에서 Gender 열을 제거 -> drop(['제거할열'], axis = 1)  axis= 0은 row를 의미 
input_data = gender_df.drop(['Gender'],axis = 1)

#제거한 Gender 열을 제외한 모든 열을 one-hot encoding 하기 
X = pd.get_dummies(input_data)
X.head()
```
